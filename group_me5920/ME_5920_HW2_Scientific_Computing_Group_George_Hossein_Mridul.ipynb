### Solution Step 2

Here, I defined functions to access the content of the input geometry and deformed geometry files. The idea is to access the files and extract the required content. Some inconsistencies in the data required replacing missing files with placeholders filled with NaN in the same shape as the expected file. Also, the data was sliced appropriately to give access to the intended content.

```python
def parse_smesh(file_path):
    """Parses a .smesh file and extracts node coordinates."""
    with open(file_path, 'r') as f:
        all_lines = f.readlines()
       
    lines = all_lines[5:-1]
    
    # If there's a 4th column, I will ignore it by slicing to [:3]
    data = np.array([list(map(float, line.split()[:3])) for line in lines])
    
    # Each .smesh is for one surface with shape (17,12,3):
    return data.reshape((17, 12, 3))


def parse_deformation(file_path):
    """Parses a deformation file with 3 surfaces (shape 17x12x3 each)."""
    try:
        with open(file_path, 'r') as f:
            all_lines = f.readlines()
    
    except FileNotFoundError:   # This is because there are a lot of missing data
       # print(f"File not found: {file_path}. Inserting placeholder.") # I commented it out because it produces a very long list
       # Return a placeholder array filled with NaN with the expected shape
        return np.full((17, 12, 3, 3), np.nan)   # Since each missing result file is to be of this shape 
        
    # Skipping the first line in the result file
    lines = all_lines[1:] 
    
    # Takes care of any 4th column, ignore it by slicing to [:3]
    data = np.array([list(map(float, line.split()[:3])) for line in lines])
    
    # Final geometry has 3 surfaces in one file → total shape (17,12,3,3)
    return data.reshape((17, 12, 3, 3))
```

### Solution Step 3

A function to load the data was defined. The function loads the files in the required shape (achieved by stacking and reshaping) and is able to access them based on timesteps, temperature, pressure and geometry, which are then stored in defined lists (and arrays).

```python
def load_data():
    """
    Loads input geometries (from smesh files) and final geometry (deformation)
    from the DM folder structure.
    """
    geometry_80_list = []  
    geometry_140_list = []  
    temp_80_list = []  
    temp_140_list = []  
    pressure_80_list = []  
    pressure_140_list = []  
    output_80_list = []
    output_140_list = []
    labels = []
    
    
    temp_values = [300, 350, 400, 450, 500]  
    pressure_values = [76, 80, 84]          
    
    for i in range(1, 65):  
        run_folder = os.path.join(INPUT_DIR, f"run{i}")
        
        smesh_files = [
            os.path.join(run_folder, "smesh.1.1.txt"),
            os.path.join(run_folder, "smesh.1.2.txt"),
            os.path.join(run_folder, "smesh.1.3.txt")
        ]
        
        if not all(os.path.exists(sf) for sf in smesh_files):
            print(f"Warning: Some smesh files missing in run{i}, skipping.")
            continue
        
        input_geometry = np.stack([parse_smesh(sf) for sf in smesh_files], axis=-1)
        
        for tstep in [80, 140]:
            for t_index in range(1, 6):  
                for p_index in range(1, 4):  
                    result_file = os.path.join(
                        FINAL_DIR,
                        f"result_{tstep}_t{t_index}_p{p_index}_{i}"
                    )
                    
                    if not os.path.exists(result_file):
                        deformation = np.full((17, 12, 3, 3), np.nan)  # Placeholder array
                    else:
                        deformation = parse_deformation(result_file)  
                    
                    temp_val = temp_values[t_index - 1]
                    press_val = pressure_values[p_index - 1]
                    
                    if tstep == 80:
                        geometry_80_list.append(input_geometry)
                        output_80_list.append(deformation)  
                        temp_80_list.append(temp_val)
                        pressure_80_list.append(press_val)
                    else:
                        geometry_140_list.append(input_geometry)
                        output_140_list.append(deformation)  
                        temp_140_list.append(temp_val)
                        pressure_140_list.append(press_val)

                    labels.append((i, temp_val, press_val))
    
    # Convert lists to NumPy arrays
    geom_80_array = np.array(geometry_80_list)
    geom_140_array = np.array(geometry_140_list)
    temp_80_array = np.array(temp_80_list)
    temp_140_array = np.array(temp_140_list)
    press_80_array = np.array(pressure_80_list)
    press_140_array = np.array(pressure_140_list)
    out_80_array = np.array(output_80_list)
    out_140_array = np.array(output_140_list)

    return geom_80_array, geom_140_array, temp_80_array, temp_140_array, press_80_array, press_140_array, out_80_array, out_140_array, labels
```

### Solution Step 4

As a result of the missing files, we cannot directly perform PCA whitening and TSNE on the output array. A mask was created in the PCA whitening function to drop those placeholder arrays before performing PCA whitening. PCA from sklearn.decomposition was used in the apply_pca_whitening() function, while TSNE from sklearn.manifold was used in the plot_tsne() function.

```python
def apply_pca_whitening(data):
    """Applies PCA whitening to normalize input and output data.
       Drops samples with NaN values before applying PCA.
    """
    from sklearn.decomposition import PCA
    reshaped = data.reshape(data.shape[0], -1)  # Flatten each sample
    # Create mask of valid samples (those with no NaNs)
    valid_mask = ~np.isnan(reshaped).any(axis=1)
    filtered_data = reshaped[valid_mask]
    dropped = reshaped.shape[0] - filtered_data.shape[0]
    if dropped > 0:
        print(f"Dropped {dropped} samples due to NaN values.")
    pca = PCA(whiten=True)
    whitened = pca.fit_transform(filtered_data)
    return whitened, valid_mask



def plot_tsne(data, labels, title):
    """Plots t-SNE results with labels based on geometry, temperature, and pressure."""
    from sklearn.manifold import TSNE
    tsne = TSNE(n_components=2, random_state=42)
    transformed = tsne.fit_transform(data)
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    label_names = ['Geometry', 'Temperature', 'Pressure']
    
    for idx, ax in enumerate(axes):
        scatter = ax.scatter(
            transformed[:, 0], transformed[:, 1],
            c=[lbl[idx] for lbl in labels],
            cmap='viridis', alpha=0.7
        )
        ax.set_title(f'{title} - {label_names[idx]}')
        fig.colorbar(scatter, ax=ax)
    
    plt.show()
```
